```
# import the necessary packages
import numpy as np
from urllib.request import urlopen
import cv2
from matplotlib import pyplot as plt

# METHOD #1: OpenCV, NumPy, and urllib
def url_to_image(url):
	# download the image, convert it to a NumPy array, and then read
	# it into OpenCV format
	resp = urlopen(url)
	image = np.asarray(bytearray(resp.read()), dtype="uint8")
	image = cv2.imdecode(image, cv2.IMREAD_COLOR)
 
	# return the image
	return image
```

```
image = url_to_image("https://fontsarena-cd5e.kxcdn.com/wp-content/uploads/2019/04/helvetica-now-font-400x364.png")
```

```
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```
edges = cv2.Canny(image,100,200)
```

```
cv2_imshow(edges)
```

```
#kernel = np.ones((3,3),np.float32)/25
kernel = np.float32([[-1,0,1],
                     [-1,0,1],
                     [-1,0,1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

```
#horizontal kernel 
#just like in vertical kernel , in this kernel we r extracting the horizontal edges by multiplying the image pixels with the below kernel.
#as this kernel multiplies with -1 on upper pixels and zero with middle ones and 1 with lower once  


kernel = np.float32([[-1,-1,-1],
                     [0,0,0],
                     [1,1,1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

```
#at 45 degree kernel
#
kernel = np.float32([[1,2,3],
                     [1,0,-1],
                     [-3,-2,-1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

```
#blur kernel 

#This kernel de-emphasize the adjacent values of the pixels , bcoz of this visibility of the image became blurred

kernel = np.float32([[0.125,0.125,0.125],
                     [0.125,0.5,0.125],
                     [-0.125,-0.125,-0.125]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)

```

```
#sharpen kernel 
#The sharpen kernel emphasizes differences in adjacent pixel values. 
#This makes the image look more vivid or sharpen.
#the diffrence between kernel values became the reson of vividness.

kernel = np.float32([[0,-2,0],
                     [-2,10,-2],
                     [0,-3,0]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

```
#identity kernel 
#only the respective pixel value will be calculated by kernel as all others ae zero.

kernel = np.float32([[0,0,0],
                     [0,1,0],
                     [0,0,0]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```
